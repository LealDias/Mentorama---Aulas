### A)	Listar nome e sobrenome ordenado por sobrenome

### C)	Liste os funcionários que têm salário superior a R$ 1.000,00 ordenados pelo nome completo

 ### F)	Liste o nome, o nome do departamento e a função de todos os funcionários

```python
sql = """
SELECT Funcionarios.PrimeiroNome, Departamentos.Nome, Funcionarios.Funcao 

FROM Funcionarios

INNER JOIN Departamentos on Departamentos.Codigo = Funcionarios.CodigoDepartamento
"""

cursor.execute(sql)
pd.DataFrame(cursor.fetchall(), columns=['Nome', 'Departamento', 'Função'])
```

 ### G)	Liste a quantidade de funcionários desta empresa

```python
sql = """
SELECT COUNT(codigo) FROM Funcionarios 
"""

cursor.execute(sql)
pd.DataFrame(cursor.fetchall(), columns= ['Soma de Funcionários'])
```

### H)	Liste o nome do departamento e do funcionário ordenados por departamento e funcionário

```python
#Ordenado por Departamento

sql = """

SELECT Departamentos.Nome, Funcionarios.PrimeiroNome

FROM Departamentos

INNER JOIN Funcionarios ON Funcionarios.CodigoDepartamento = Departamentos.Codigo

ORDER BY Departamentos.Nome

"""

cursor.execute(sql)
pd.DataFrame(cursor.fetchall(), columns= ['Departamento', 'Funcionário'])
```

```python
#Ordenado por Funcionários
sql = """

SELECT Departamentos.Nome, Funcionarios.PrimeiroNome

FROM Departamentos

INNER JOIN Funcionarios ON Funcionarios.CodigoDepartamento = Departamentos.Codigo

ORDER BY Funcionarios.PrimeiroNome

"""

cursor.execute(sql)
pd.DataFrame(cursor.fetchall(), columns= ['Departamento', 'Funcionário'])
```
