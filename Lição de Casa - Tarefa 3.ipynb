```python
class Quadrado:
    
    def __init__(self, tam_lado): #Método Construtor
        
        self.tam_lado = tam_lado                #A)
        
        
    def Muda_Lado(self, tam_lado):  #Método Setter            #B)
        
        self.tam_lado = tam_lado
        
    def Retorna_Lado(self): #Método Getter
        
        return self.tam_lado
    
    def Cal_Area(self): # Método Getter
        
        return self.tam_lado * self.tam_lado
    
```

## EXERCÍCIO 3

```python
class Retangulo:
    
    def __init__(self, base, altura): # Método Construtor   #A)
        
        self.base = base
        self.altura = altura
        
        
    def Mudar_Valor(self, base, altura):  #Setter            #B)
        
        self.base = base
        self.altura = altura
        
    def Retorna_Lados(self):              #Getter
        
        return self.base, self.altura
    
    def Calc_Area(self):                  #Getter
        
        return self.base * self.altura
    
    def Calc_Per(self):                   #Getter
        
        return 2*(self.base + self.altura)
        
```

```python
class Pessoa:
    
    def __init__(self, nome, idade, peso, altura): #Método Construtor        #A)
        
        self.nome = nome
        self.idade = idade
        self.peso = peso
        self.altura = altura
        
                                                                              #B)
    
    def Envelhecer(self, idade):
        
        self.idade = self.idade + idade
        
        if self.idade < 21:
            
            self.altura = self.altura + (idade * 0.05)
        
        
        return self.idade

            
    def Engordar(self, peso):
        
        self.peso = self.peso + peso
        
        print("%s está engordando...seu peso agora é: %.fKg" %(self.nome, float(self.peso)))
        
        
    def Emagrecer(self, peso):
        
        self.peso = self.peso - peso
        
        print("%s está emagrecendo...seu peso Agora é: %.f Kg" % (self.nome, float(self.peso)))
        
        
    def Crescer(self): # Mostra a altura de acordo com a idade inserida na def Envelhecer
        
        if self.idade < 21:
            
            print("A altura de %s é: %.2f Metros" % (self.nome, float(self.altura)))
            
        else:
                
            print("Mais do que 21 anos...Sem crescimento")
            
        
        
        
  
    

```

```python
class Carro:
    
    def __init__(self, cor, ano, qty_portas): #Método Construtor
        
        self.cor = cor
        self.ano = ano
        self.qty_portas = qty_portas
        
    def Imprime_Carro(self):
        
        print("Cor do Carro: %s" %self.cor)
        print("Ano do Carro: %d" %self.ano)
        print("Quantidade de Portas: %d" %self.qty_portas)
        
        
    def Mostra_Cor(self):
        
        return self.cor
    
        
    def Mostra_Ano(self):
        
        return self.ano
      
```

## EXERCÍCIO 7

```python
class TV:
    
    def __init__(self, canal = 0, volume = 0):
        
        self.canal = canal
        self.volume = volume
        
    def Muda_Canal(self, canal):
        
        if canal in range(1,20):
            
            self.canal = canal
            
            return self.canal
            
        else:
                
            print("Fora da Faixa de Canais. Digite um Valor de 1 a 20")
            
            
    def Aumenta_Volume(self, volume):
        
        
        if (self.volume + volume) in range(1,40):
            
            self.volume = self.volume + volume
            
            return self.volume
            
        else:
            
            print("Fora da Faixa de Volumes. O faixa de valores é de 0 a 40")
            
            
    def Diminui_Volume(self, volume):
        
        if (self.volume - volume) in range(1,40):
            
            self.volume = self.volume - volume
            
            return self.volume
        
        else:
            
             print("Fora da Faixa de Volumes. O faixa de valores é de 1 a 40")
            
            
    
```

```python
#--------------------------------------------------------------------------------------------------------------#

#Programa Principal

televisao = TV()

while True:
    
    print("TV Ligada...")
    
    opc = int(input(("Qual ação deseja tomar?" "\n" "1 - Trocar de Canal " "\n" "2 - Mudar Volume" "\n" "3 - Desligar TV" "\n")))
    
    if opc == 1:
    
        canal = int(input("Digite um canal válido (1 a 20): "))
        print("Canal Atual: " , televisao.Muda_Canal(canal))
        
        
    elif opc == 2:
        
            opc_2 = int(input("Comandos: 1 - Aumentar(+)" "\n" "2 - Diminuir(-)" "\n"))
        
            if opc_2 == 1:
        
                volume = int(input("Digite um Volume (Max Vol: 40): "))
                print("Voulme Atual" , televisao.Aumenta_Volume(volume))
                
                
            elif opc_2 == 2:
                
                volume = int(input("Digite um Volume (Max Vol: 40): "))
                print("Volume Atual" , televisao.Diminui_Volume(volume))
            
        
    elif opc == 3:
        
        print("Desligando TV...")
        
        break
            
            
```

## EXERCÍCIO 8

```python
class Bomba_de_Combustivel:
    
    def __init__(self, tipoCombustivel = " ", valorLitro = 0, quantCombustivel = 0): # Método Construtor       #A)
        
        self.tipoCombustivel = tipoCombustivel
        self.valorLitro = valorLitro
        self.quantCombustivel = quantCombustivel
        
        
    def abastecePorValor(self, valor):                                                                        #B)
        
        litros_abastecidos = (valor / self.valorLitro) * 1     #-> 1 é a representação do litro
        
        self.quantCombustivel = self.quantCombustivel - litros_abastecidos
        
        return litros_abastecidos
    
    
    def abastecePorLitro(self, litros):
        
        self.quantCombustivel = self.quantCombustivel - litros
        
        return self.valorLitro * litros
        
    
    def alteraValor(self, valor):
        
        self.valorLitro = valor
        
        print("Valor do Litro Alterado com Sucesso para %.2f" % self.valorLitro)
    
    
    def alteraCombustivel(self, tipo):
        
        self.tipoCombustivel = tipo
        
        print("Tipo de Combustível Alterado para: %s" %self.tipoCombustivel)
        
    
    def alteraQtyCombustivel(self, quantidade):
        
        self.quantCombustivel = quantidade
        
        print("Quantidade de Combustível Atualizada para: %.2f" %self.quantCombustivel)
        
    def ver_Relatorio(self):
        
        print("--------------------INFORMAÇÕES DE CADASTRO---------------------------")
        print("TIPO DE COMBUSTÍVEL: ", self.tipoCombustivel)
        print("PREÇO DO COMBUSTÍVEL %.2f: " %self.valorLitro)
        print("TOTAL NO RESERVATÓRIO: ", self.quantCombustivel)
        print("-----------------------------------------------------------------------")
        
        

```

```python

print("--------------------INFORMAÇÕES DE CADASTRO---------------------------")

tipo = str(input("Tipo de Combustível: "))
valor = float(input("Preço do Combustível: "))
qty_Comb = float(input("Quantidade nos Reservatórios em Litros: "))

posto = Bomba_de_Combustivel(tipo, valor, qty_Comb) # Instância da Classe 

print("-----------------------------------------------------------------------")

while True:
    
    val = 0
    lit = 0
    
    
    print("Bem Vindo ao Posto de Combustíveis MENTORAMA!!")
    
    
    opc = int(input("O que deseja fazer?" "\n"
                    "1 - ABASTECER" "\n"
                    "2 - ALTERAR DADOS DE CADASTRO" "\n"
                    "3 - VER RELATÓRIO" "\n"
                    "4 - SAIR DO POSTO" "\n"
                    ))
    
#-----------------------------------------------------------------------------------------------------#
    if opc == 1:
        
        opc_1 = int(input("Deseja abastecer pelo valor ou por litro?" "\n" " 1 - VALOR" "\n" " 2 - LITRO"))
        
        if opc_1 == 1:
            
            val = float(input("Qual será o valor?: "))
            print("Ok... Serão R$ %.2f" %val)
            
            print("Foram abastecidos %.2f Litros \n"  %posto.abastecePorValor(val))
            
        elif opc_1 == 2:
            
            lit = float(input("Quantos Litros?: "))
            print("Ok... Serão %.2f Litros" %lit)
            
            print("Foram Gastos R$ %.2f \n"  %posto.abastecePorLitro(lit))
#-----------------------------------------------------------------------------------------------------#            
            
    if opc == 2:
        
        posto.alteraCombustivel(str(input("Tipo de Combustível: ")))
        posto.alteraValor(float(input("Preço do Combustível: ")))
        posto.alteraQtyCombustivel(float(input("Quantidade nos Reservatórios em Litros: \n")))
        
        print("Dados Alterados com Sucesso!!! \n")
        
#---------------------------------------------------------------------------------------------------#  

    if opc == 3:
        
        
        posto.ver_Relatorio()
        
#---------------------------------------------------------------------------------------------------#        
     
    elif opc == 4:  #Flag de Saída...
        
        print("Obrigado por Abastecer Conosco...!!!!! \n")
        break
        
        
        
    
```
