```python
# Resposta questão 5a: 

sex = df.Sexo.value_counts()

x = sex[0]
y = sex[1]

nomes = ['Homens', 'Mulheres']
val = [x,y]

plt.bar(nomes, val, color =['maroon', 'r'], label = "Frequência de Reclamação por Sexo", width = 0.4)
plt.legend()
plt.xlabel('Sexo')
plt.ylabel('Quantidade de Reclamações')
plt.show()
```

```python
# Resposta questão 5b: 

UF = df.UF.value_counts()

valores = list()
nomes = list()
#------------------------------------------------------------------------------------------------------------#
for i in UF:
    
    valores.append(i)

for j in UF.keys():
    
    nomes.append(j)
#------------------------------------------------------------------------------------------------------------#    
fig = plt.figure(figsize = (15, 5)) #Estabelece um tamanho maior para que as Letras caibam no Gráfico
    
plt.bar(nomes, valores, color = 'r', label = "Frequência de Reclamação por Estado", width = 0.6)
plt.legend()
plt.xlabel('Estados')
plt.ylabel('Quantidade de Reclamações')
plt.show()
```

```python
# Resposta questão 5c: 

respondida = df['Respondida'].value_counts()[1] # Não respondidos
nao_respondida = df['Respondida'].value_counts()[0] # Respondidas

nomes = ['Não Respondida', 'Respondidas']
valores = [respondida, nao_respondida]

fig = plt.figure(figsize = (10, 5)) #Estabelece um tamanho maior para que as Letras caibam no Gráfico
    
plt.bar(nomes, valores, color = 'r', label = "Respondidas vs Não Respondidas", width = 0.6)
plt.legend()
plt.xlabel('Status da Reposta')
plt.ylabel('Quantidade de Reclamações')
plt.show()
```
